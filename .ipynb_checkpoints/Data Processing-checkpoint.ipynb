```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns

import plotly.express as pl
```

```python
data = pd.read_csv('seasons_stats.csv')
data.head()
```

```python
#Only run this once
data.drop(['Unnamed: 0', 'blanl', 'blank2'],axis = 1, inplace = True)
data.dropna(inplace = True)
data['Year'] = data['Year'].astype('int32')


enc_df = pd.get_dummies(data[['Pos']])
# merge with main df bridge_df on key values
data = data.join(enc_df)
data.drop("Pos",axis = 1, inplace = True)
```

```python
data.describe()
```

```python
fig = pl.histogram(data, x="PER")
fig.show()
```

```python
df = data.groupby(['Year','Player']).agg('mean')
df.duplicated().sum()
df
```

```python
per = df
per = per.iloc[:,per.columns == 'PER']
per = per['PER'].unstack().shift(+1) # pull out the groups, shift with lag step=1
per = per.stack(dropna=True);
per
```

```python
merged = pd.concat([df, per],axis = 1, join = 'outer')
merged.dropna(inplace = True)
merged.reset_index(inplace = True)
merged.rename(columns = {0 : 'PER-1'}, inplace = True)
merged.sort_values(['Player','Year'])
```
